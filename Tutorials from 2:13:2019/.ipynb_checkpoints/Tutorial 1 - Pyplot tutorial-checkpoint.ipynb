```python
import matplotlib.pyplot as plt
import numpy as np #will be used later in tutorial
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

```python
#plot() will take arbitrary number of arguments - if only given one array (like above) will assume that is y and generate x vals
#'ro' indicates color and shape of points on plot
plt.plot([1,2,3,4], [1,4,9,16], 'ro') 

#axis() command takes list of args in form of [xmin, xmax, ymin, ymax]
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
#use numpy to evenly sample time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
#several ways to control line properties shown below
x = [1,2,3,4]
y = [1,4,9,16]

#use keyword args
plt.plot(x, y, linewidth=2.0)
print(plt.show())

#use setup function
lines = plt.plot(x,y) #(x1,y1,x2,y2)
#use setp function to set options related to plot
plt.setp(lines, color='r', linewidth=2.0)
print(plt.show())
```

```python
#Create two subplots

#helper function to 
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

#working with figure 1
plt.figure(1)
#subplot with identifier (still on figure 1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

#subplot wiht identifier (also still on figure 1)
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()

#subplot() function takes argument in form numrows,numcols,fignum
```

```python
#creating multiple figures with consecutive calls to figure()

plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
plt.show()
```

```python
#text() command can be used to insert text at an arbitrary location on the plot as shown below


# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') #can see that the indices correspond to position on graph
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
#Annotate function provides interface for adding annotations to the graph

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            ) #note that xy and xytext dictate the position pointed to and the position of the text used to annotate

plt.ylim(-2,2)
plt.show()
```

```python
#Can also change the scale of a plots axis relatively easily

from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
