```python
#import necessary packages
import warnings
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
season_data = pd.read_csv("Team_Records.csv")
```

```python
#simple graph of celtics wins per season over time

#get celtics data from season data
celtics_data = season_data[season_data.Team == "Boston Celtics*"]

#modify season column so that it only reflects the start of the season
celtics_data['Season'] = pd.to_numeric(celtics_data['Season'].str[:4])
plt.plot(celtics_data['Season'],celtics_data['W'],'-ok')
plt.show()
```
